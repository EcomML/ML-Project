```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

rc('font', family='Malgun Gothic')

male_cust = user_data[user_data['Gender'] == 'Male']
female_cust = user_data[user_data['Gender'] == 'Female']

def categorize_age(age):
    if age >= 10 and age < 20:
        return '10대'
    elif age >= 20 and age < 30:
        return '20대'
    elif age >= 30 and age < 40:
        return '30대'
    elif age >= 40 and age < 50:
        return '40대'
    elif age >= 50 and age < 60:
        return '50대'
    else:
        return '고연령층'
    
# 나이대 컬럼 추가
user_data['Age_cat'] = user_data['Age'].apply(categorize_age)
male_cust['Age_cat'] = male_cust['Age'].apply(categorize_age)
female_cust['Age_cat'] = female_cust['Age'].apply(categorize_age)

# null 값이 있는지 확인 
user_data['Age_cat'].isnull().sum
male_cust['Age_cat'].isnull().sum()
female_cust['Age_cat'].isnull().sum()

# 파이 차트 내 글씨
def func(pct, allvalues):
    absolute = int(pct/100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} 명)".format(pct, absolute)

# 전체 고객의 나이대 분포 파이 차트
age_cat_counts_total = user_data['Age_cat'].value_counts()

plt.figure(figsize=(21, 7))
plt.subplot(1, 3, 1)
plt.pie(age_cat_counts_total, labels=age_cat_counts_total.index, autopct=lambda pct: func(pct, age_cat_counts_total), startangle=140,colors=sns.color_palette('Greens', len(age_cat_counts_total)))
plt.title('Age Distribution of all Customers')

# 남성 고객의 나이대 분포 파이 차트
age_cat_counts_male = male_cust['Age_cat'].value_counts()

plt.subplot(1, 3, 2)
plt.pie(age_cat_counts_male, labels=age_cat_counts_male.index, autopct=lambda pct: func(pct, age_cat_counts_male), startangle=140,colors=sns.color_palette('Blues', len(age_cat_counts_male)))
plt.title('Age Distribution of male Customers')

# 여성 고객의 나이대 분포 파이 차트
age_cat_counts_female = female_cust['Age_cat'].value_counts()
plt.subplot(1, 3, 3)
plt.pie(age_cat_counts_female, labels=age_cat_counts_female.index, autopct=lambda pct: func(pct, age_cat_counts_female), startangle=140,colors=sns.color_palette('Reds', len(age_cat_counts_female)))
plt.title('Age Distribution of female Customers')
plt.tight_layout()
plt.show()
```

```python
# 남성과 여성 데이터로 나누기
male_cust = user_data[user_data['Gender'] == 'Male']
female_cust = user_data[user_data['Gender'] == 'Female']

# 남성과 여성 각각에서 나이대별로 구매된 품목 빈도 계산
male_item_counts = male_cust.groupby(['Age_cat', 'Item Purchased'])['Customer ID'].count().reset_index()
female_item_counts = female_cust.groupby(['Age_cat', 'Item Purchased'])['Customer ID'].count().reset_index()

# 남성과 여성 각각에서 나이대별로 가장 많이 구매된 품목 찾기
male_top_items = male_item_counts.loc[male_item_counts.groupby('Age_cat')['Customer ID'].idxmax()]
female_top_items = female_item_counts.loc[female_item_counts.groupby('Age_cat')['Customer ID'].idxmax()]

# 시각화: 남성과 여성 각각에서 나이대별로 가장 많이 구매된 품목 표시
fig, axs = plt.subplots(1, 2, figsize=(16, 8))

# 남성 그래프
for age_cat in male_top_items['Age_cat'].unique():
    data = male_top_items[male_top_items['Age_cat'] == age_cat]
    axs[0].bar(data['Item Purchased'], data['Customer ID'], label=age_cat)
axs[0].set_title('남성 고객의 나이대별 주요 구매 품목')
axs[0].set_xlabel('구매 품목')
axs[0].set_ylabel('구매 빈도')
axs[0].legend()

# 여성 그래프
for age_cat in female_top_items['Age_cat'].unique():
    data = female_top_items[female_top_items['Age_cat'] == age_cat]
    axs[1].bar(data['Item Purchased'], data['Customer ID'], label=age_cat)
axs[1].set_title('여성 고객의 나이대별 주요 구매 품목')
axs[1].set_xlabel('구매 품목')
axs[1].set_ylabel('구매 빈도')
axs[1].legend()

plt.tight_layout()
plt.show()
```

```python
# 남성 고객만 선택
male_data = user_data[user_data['Gender'] == 'Male']

# 계절별, 연령대별로 가장 많이 구매된 품목 찾기
season_age_item_counts_male = male_data.groupby(['Season', 'Age_cat'])['Item Purchased'].value_counts().unstack().fillna(0)

# 계절별 연령대별 추천 품목 찾기 (가장 많이 구매한 품목)
season_age_recommendations_male = season_age_item_counts_male.idxmax(axis=1).unstack()

# 시각화를 위한 데이터 준비
recommendations_data_male = season_age_recommendations_male.reset_index().melt(id_vars=['Season'], var_name='Age_cat', value_name='Recommended Item')

# 계절별, 연령대별 추천 품목 시각화
fig, axes = plt.subplots(2, 2, figsize=(15, 12), sharex=True)

# 계절 순서 정의
seasons = ['Spring', 'Summer', 'Fall', 'Winter']

for i, season in enumerate(seasons):
    ax = axes[i // 2, i % 2]
    season_data = recommendations_data_male[recommendations_data_male['Season'] == season]
    bars = sns.barplot(x='Age_cat', y=season_data.index, data=season_data, ax=ax, palette='pastel')

    # 막대에 레이블 추가
    for bar, (idx, row) in zip(bars.patches, season_data.iterrows()):
        ax.text(bar.get_x() + bar.get_width() / 2, bar.get_y() + bar.get_height() / 2,
                row['Recommended Item'], ha='center', va='center', fontsize=10, color='black')

    ax.set_title(f'{season} - Age Category')
    ax.set_xlabel('Age Category')
    ax.set_ylabel('Recommended Item')
    ax.tick_params(axis='x', rotation=45)

plt.suptitle('Seasonal Recommended Items by Age Category (Male)', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
# 여성 고객만 선택
female_data = user_data[user_data['Gender'] == 'Female']

# 계절별, 연령대별로 가장 많이 구매된 품목 찾기
season_age_item_counts = female_data.groupby(['Season', 'Age_cat'])['Item Purchased'].value_counts().unstack().fillna(0)

# 계절별 연령대별 추천 품목 찾기 (가장 많이 구매한 품목)
season_age_recommendations = season_age_item_counts.idxmax(axis=1).unstack()

# 시각화를 위한 데이터 준비
recommendations_data = season_age_recommendations.reset_index().melt(id_vars=['Season'], var_name='Age_cat', value_name='Recommended Item')

# 계절별, 연령대별 추천 품목 시각화
fig, axes = plt.subplots(2, 2, figsize=(15, 12), sharex=True)

# 계절 순서 정의
seasons = ['Spring', 'Summer', 'Fall', 'Winter']

for i, season in enumerate(seasons):
    ax = axes[i // 2, i % 2]
    season_data = recommendations_data[recommendations_data['Season'] == season]
    bars = sns.barplot(x='Age_cat', y=season_data.index, data=season_data, ax=ax, palette='pastel')

    # 막대에 레이블 추가
    for bar, (idx, row) in zip(bars.patches, season_data.iterrows()):
        ax.text(bar.get_x() + bar.get_width() / 2, bar.get_y() + bar.get_height() / 2,
                row['Recommended Item'], ha='center', va='center', fontsize=10, color='black')

    ax.set_title(f'{season} - Recommended Item by Age Category')
    ax.set_xlabel('Age Category')
    ax.set_ylabel('Recommended Item')
    ax.tick_params(axis='x', rotation=45)

plt.suptitle('Seasonal Recommended Items by Age Category (Female)', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

## 남성 고객의 나이대별 구매 관여도

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 나이대 컬럼 추가
male_data['Age_cat'] = male_data['Age'].apply(categorize_age)

# 나이대별 구매 빈도 및 총 구매 금액 계산
age_purchase_counts = male_data['Age_cat'].value_counts().sort_index()
age_purchase_sums = male_data.groupby('Age_cat')['Purchase Amount (USD)'].sum()

# 나이대별 구매한 품목 분석
age_item_counts = male_data.groupby('Age_cat')['Item Purchased'].value_counts().unstack().fillna(0)
# 나이대별 구매한 품목 수를 많이 구매한 순서대로 정렬
sorted_age_item_counts = age_item_counts.sum(axis=1).sort_values(ascending=False)

# 구매 빈도 및 총 구매 금액 시각화
fig, ax1 = plt.subplots(figsize=(14, 8))

color = 'tab:blue'
ax1.set_xlabel('나이대')
ax1.set_ylabel('구매 빈도', color=color)
sns.barplot(x=age_purchase_counts.index, y=age_purchase_counts.values, ax=ax1, palette='Blues_d')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  
color = 'tab:red'
ax2.set_ylabel('총 구매 금액 (USD)', color=color)
sns.lineplot(x=age_purchase_sums.index, y=age_purchase_sums.values, ax=ax2, color=color, marker='o')
ax2.tick_params(axis='y', labelcolor=color)

plt.title('남성 고객의 나이대별 구매 패턴')
fig.tight_layout()  
plt.show()

# 나이대별 구매한 품목 시각화
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 12))
axes = axes.flatten()

# 나이대별로 구매한 품목을 많이 구매한 순서대로 시각화
for i, age_cat in enumerate(sorted_age_item_counts.index):
    ax = axes[i]
    data = age_item_counts.loc[age_cat].sort_values(ascending=False)  # 품목 수를 많이 구매한 순서대로 정렬
    data.plot(kind='bar', ax=ax, color=sns.color_palette('pastel', len(data)))
    ax.set_title(f'{age_cat} - 구매한 품목 분포')
    ax.set_xlabel('Item Purchased')
    ax.set_ylabel('Count')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45)  

# 10번 인덱스부터 서브플롯 보이기
for j in range(10, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
# 카테고리별 상품별 평균 구매 금액 계산
category_item_avg_purchase = user_data.groupby(['Category', 'Item Purchased'])['Purchase Amount (USD)'].mean().unstack()

# 시각화를 위한 준비
categories = category_item_avg_purchase.index.unique()
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 파이 차트 내 글씨 흰색으로 설정하는 함수
def func(pct, allvalues):
    absolute = int(pct/100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} USD)".format(pct, absolute)

# 각 카테고리별로 파이 차트 그리기
for ax, category in zip(axes.flatten(), categories):
    # 해당 카테고리의 상품만 필터링
    category_data = category_item_avg_purchase.loc[category].dropna()
    category_data = category_data.sort_values(ascending=False)
    wedges, texts, autotexts = ax.pie(category_data, labels=category_data.index, autopct=lambda pct: func(pct, category_data), startangle=140, colors=sns.color_palette('viridis', len(category_data)))
    ax.set_title(f'{category} - 상품별 평균 구매 금액')
    for text in texts:
        text.set_color('black')
    for autotext in autotexts:
        autotext.set_color('white')

plt.tight_layout()
plt.show()
```
